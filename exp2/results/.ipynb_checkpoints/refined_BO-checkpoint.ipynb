```python
import numpy as np
import glob
import pickle
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import sys
sys.path.append('../code/')
%matplotlib inline  
from bayesian_optimisation import BayesOptParticleFiltering
```

```python
N = 100
paths = glob.glob('res/N_100/bayesopt_*')
print('found {0} files'.format(len(paths)))
n_simul         = 50
max_lkd         = np.zeros([2, n_simul])
max_lkd_recover = np.zeros([2, n_simul])
param           = np.zeros([2, n_simul, 3])
low_b           = np.array([0., 0., 0.], dtype=np.double)
upp_b           = np.array([1., 1., 1.], dtype=np.double) 
for s_idx in range(1, n_simul + 1):
    try:
        p_0 = glob.glob('res/N_100/bayesopt_{0}_refine0.pkl'.format(s_idx))[0]
        i_0 = pickle.load(open(p_0, 'rb'))
        max_lkd[0, s_idx - 1]     = -i_0[0]
        param[0, s_idx - 1]       = .5 * (10**i_0[1] - 10**0.)/(10**1. - 10**0.) 
        p_1 = glob.glob('res/N_100/bayesopt_{0}_refine1.pkl'.format(s_idx))[0]
        i_1 = pickle.load(open(p_1, 'rb'))
        max_lkd[1, s_idx - 1]     = np.log(-i_1[0])
        param[1, s_idx - 1]       = .5 * (10**i_1[1] - 10**0.)/(10**1. - 10**0.)
        bayesOptParticleFiltering = BayesOptParticleFiltering(s_idx, low_b, upp_b, nb_it=100, nb_ini=100, refine=False, load_save_flag=2)
        max_lkd_recover[0, s_idx - 1] = -bayesOptParticleFiltering.evaluateSample(i_0[1])
        max_lkd_recover[1, s_idx - 1] = -bayesOptParticleFiltering.evaluateSample(i_1[1])        
    except IndexError:
        print('error with subject {0}'.format(s_idx))
max_lkd = np.delete(max_lkd, 27, axis=1)
param   = np.delete(param, 27, axis=1)
max_lkd_recover = np.delete(max_lkd_recover, 27, axis=1)
```

```python
fig=plt.figure(figsize=(12,4))
from scipy.stats import gaussian_kde

plt.suptitle('Parameters, N=100', fontsize=20)

l1, l2       = 'Refined', 'Brute'
red_patch  = mpatches.Patch(color='red', label=l1)
blue_patch = mpatches.Patch(color='blue', label=l2)
fig.legend((red_patch, blue_patch),(l1, l2),loc=1, fontsize=10,ncol=2)

ax = plt.subplot(131)
theta_idx = 0
density1, density2 = gaussian_kde(param[1,:,theta_idx]), gaussian_kde(param[0,:,theta_idx]) #, gaussian_kde(params['00101'][0]), gaussian_kde(params['00001'][0])
xs = np.linspace(0,.5,1000)
density1.covariance_factor = lambda : .45; density2.covariance_factor = lambda : .45; #density3.covariance_factor = lambda : .45
density1._compute_covariance(); density2._compute_covariance(); #density3._compute_covariance()
ax.plot(xs, density1(xs), 'r')
ax.fill_between(xs, 0, density1(xs), color='red', alpha=.2)
ax.plot(xs, density2(xs), 'b')
ax.fill_between(xs, 0, density2(xs), color='blue', alpha=.2)
ax.set_ylabel('density')
ax.set_xlabel('r')

ax = plt.subplot(132)
theta_idx = 1
density1, density2 = gaussian_kde(param[1,:,theta_idx]), gaussian_kde(param[0,:,theta_idx]) #, gaussian_kde(params['00101'][0]), gaussian_kde(params['00001'][0])
xs = np.linspace(0,.5,1000)
density1.covariance_factor = lambda : .45; density2.covariance_factor = lambda : .45; #density3.covariance_factor = lambda : .45
density1._compute_covariance(); density2._compute_covariance(); #density3._compute_covariance()
ax.plot(xs, density1(xs), 'r')
ax.fill_between(xs, 0, density1(xs), color='red', alpha=.2)
ax.plot(xs, density2(xs), 'b')
ax.fill_between(xs, 0, density2(xs), color='blue', alpha=.2)
ax.set_ylabel('density')
ax.set_xlabel('sigma')

ax = plt.subplot(133)
theta_idx = 2
density1, density2 = gaussian_kde(param[1,:,theta_idx]), gaussian_kde(param[0,:,theta_idx]) #, gaussian_kde(params['00101'][0]), gaussian_kde(params['00001'][0])
xs = np.linspace(0,.5,1000)
density1.covariance_factor = lambda : .45; density2.covariance_factor = lambda : .45; #density3.covariance_factor = lambda : .45
density1._compute_covariance(); density2._compute_covariance(); #density3._compute_covariance()
ax.plot(xs, density1(xs), 'r')
ax.fill_between(xs, 0, density1(xs), color='red', alpha=.2)
ax.plot(xs, density2(xs), 'b')
ax.fill_between(xs, 0, density2(xs), color='blue', alpha=.2)
ax.set_ylabel('density')
ax.set_xlabel('tau')
```

```python
fig=plt.figure(figsize=(12,4))

plt.suptitle('Re-estimated Likelihoods, N=100', fontsize=15)
mean_lkd_reestimated = np.mean(max_lkd_recover, axis=1)
err_lkd_reestimated  = np.std(max_lkd_recover, axis=1)/np.sqrt(n_simul)
ax = plt.subplot(111)
ax.bar(0., mean_lkd_reestimated[0], yerr=err_lkd_reestimated[0], width=0.6, color='b', ecolor='k', align='center')
ax.bar(1., mean_lkd_reestimated[1], yerr=err_lkd_reestimated[1],  width=0.6, color='r', ecolor='k', align='center')
label = ['', '', 'Brute', '', '', '', '', 'Refined']
ax.set_xticklabels(label)
ax.set_ylim((30,50))
```
